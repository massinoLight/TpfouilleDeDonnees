# A

```python
from sklearn import *
import numpy as np
# 1- Creer une matrice X 
X = np.array([[1,-1,2],[2,0,0],[0,1,-1]])
print(X)
```

```python
# 2- La moyenne de X et la variance 
print(X.mean())
print(X.var())
```

```python
# 3- Normaliser la matrice avec scale
from sklearn.preprocessing import scale

Xnorm=scale(X)
print(Xnorm)
#On remarque qu'il y a des valeurs >1 et des valeurs <1 
```

```python
# 4- Calculer la moyenne et la variance de la matrice normalisé 
print(Xnorm.mean())
print(Xnorm.var())
#La variance d'une matrice normalisé est toujours 1
```

```python
# 1- Creer une matrice X2
X2 = np.array([[1,-1,2],[2,0,0],[0,1,-1]])
print(X2)
print()
# 2- Calculer la moyenne et la variance sur les variables  
print(X2.mean(0))
print(X2.var(0))
```

```python
print(X2norm.mean(0))
print(X2norm.var(0))
#On constate que les valeurs obtenu apres normalisation sont plus cohérentes et parlantes 
```

# C visualisation de données

```python
from sklearn import *
print("# 1 - Chargez les données  Iris:")
iris = datasets.load_iris()
```

# Visualiser le nuage de points en utilisant toutes les combinaisons possibles CNP=n!/p!(n-p)! == 4!/3! =4

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
data=iris.data

target=iris.target
x1 = []; y1 = []; z1 = []
x2 = []; y2 = []; z2 = []
x3 = []; y3 = []; z3 = []
for i in range(0,len(data)):
        if target[i]==0:
           x1.append(data[i][0])
           y1.append(data[i][1])
           z1.append(data[i][2])
        else:
            if target[i]==1:
               x2.append(data[i][0])
               y2.append(data[i][1])
               z2.append(data[i][2])
            else:
                if target[i]==2:
                    x3.append(data[i][0])
                    y3.append(data[i][1])
                    z3.append(data[i][2])
                   
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')





ax.scatter(x1, y1, z1, c='r', marker='o')
ax.scatter(x2, y2, z2, c='b', marker='o')
ax.scatter(x3, y3, z3, c='g', marker='o')

ax.set_xlabel('sepal length(cm)')
ax.set_ylabel('sepal width (cm)')
ax.set_zlabel('petal length (cm)')

plt.show()               
```

#visualisation en 2D

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)



plt.figure()
colors = ['blue', 'red', 'green']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')

plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')

plt.show()
```

On constate que les classes verginica et versicolor sont plus séparé lors de LinearDiscriminantAnalysis 

LDA, contrairement à PCA, est une méthode supervisée, utilisant des étiquettes de classe connues.
