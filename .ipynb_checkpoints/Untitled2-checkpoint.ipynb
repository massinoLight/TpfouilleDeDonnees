```python
# 3 - Afficher le nom des classes pour chaque données
print("# 3 - Afficher le nom des classes pour chaque données")
data=iris.data
print(type(data))
target=iris.target
for i in range(0,len(data)):
        if target[i]==0:
           print(data[i],"setosa")
        else:
            if target[i]==1:
                print(data[i], "versicolor")
            else:
                if target[i]==2:
                    print(data[i], "virginica")
```

```python
from matplotlib import pyplot
from sklearn import *
import numpy
print("# 1 - Afficher l'aide pour utiliser la fonction datasets.make_blobs ")
print(help(datasets.make_blobs))
print("")
print("# 2 - Génerer 1000 données de deux varibables réparties en 4 groupes ")
randomdata=datasets.make_blobs(n_samples=1000, n_features=2, centers=4)
print(randomdata)
print("")
X,Y=datasets.make_blobs(n_samples=1000, n_features=2, centers=4)
```

```python
# 3 - Afficher les données avec des couleurs
pyplot.figure(figsize = (10, 10))
pyplot.title("le super titre ")
pyplot.xlim(-15,15)
pyplot.ylim(-15,15)
pyplot.xlabel("laxe des X")
pyplot.ylabel("laxe des Y")
x1 = []; y1 = []
x2 = []; y2 = []
x3 = []; y3 = []
x4 = []; y4 = []
for i in range(0,len(X)):
    if Y[i] == 0:
        x1.append(X[i][0])
        y1.append(X[i][1])
    else:
        if Y[i] == 1:
            x2.append(X[i][0])
            y2.append(X[i][1])
        else:
            if Y[i] == 2:
                x3.append(X[i][0])
                y3.append(X[i][1])
            else:
                if Y[i] == 3:
                    x4.append(X[i][0])
                    y4.append(X[i][1])





pyplot.scatter(x1, y1, c = 'red')
pyplot.scatter(x2, y2, c = 'yellow')
pyplot.scatter(x3, y3, c = 'blue')
pyplot.scatter(x4, y4, c = 'green')
pyplot.show()
```

```python
from matplotlib import pyplot
from sklearn import *
import numpy
#4 -Concatiner deux jeux de données
X1,Y1=datasets.make_blobs(n_samples=100, n_features=2, centers=2)
X2,Y2=datasets.make_blobs(n_samples=500, n_features=2, centers=3)

conc1=numpy.vstack((X1,X2))
conc2=numpy.hstack((Y1,Y2))
```

```python
pyplot.figure(figsize = (10, 10))
pyplot.title("le super titre ")
pyplot.xlim(-15,15)
pyplot.ylim(-15,15)
pyplot.xlabel("laxe des X")
pyplot.ylabel("laxe des Y")
x1 = []; y1 = []
x2 = []; y2 = []
x3 = []; y3 = []
x4 = []; y4 = []
for i in range(0,len(conc1)):
    if conc2[i] == 0:
        x1.append(conc1[i][0])
        y1.append(conc1[i][1])
    else:
        if conc2[i] == 1:
            x2.append(conc1[i][0])
            y2.append(conc1[i][1])
        else:
            if conc2[i] == 2:
                x3.append(conc1[i][0])
                y3.append(conc1[i][1])






pyplot.scatter(x1, y1, c = 'red')
pyplot.scatter(x2, y2, c = 'yellow')
pyplot.scatter(x3, y3, c = 'blue')

pyplot.show()
```
